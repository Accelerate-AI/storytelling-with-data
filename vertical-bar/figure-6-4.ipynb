```python
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np

#%matplotlib inline
%matplotlib
```

```python
# configure plot font family to Arial
plt.rcParams['font.family'] = 'Arial'
# configure mathtext bold and italic font family to Arial
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'

X = ['Today\n9/30/15', 'FY16', 'FY17', 'FY18', 'FY19', 'FY20']
YA = np.array([0, -28, -44, -62, -77, -99])
YT = np.array([190, 162, 146, 128, 113, 91])
YP = np.array([0, 7, 10, 14, 16, 22])
YQ = np.array([0, 4, 4, 4, 4, 4])
YU = np.array([0, 28, 50, 68, 91, 112])

# Silver
color_gray1 = '#BFBEBE'
# Smoky
color_gray4 = '#646369'
# Aluminium
color_gray5 = '#929497'
# Nero
color_gray6 = '#231F20'
# Polo Blue
color_blue1 = '#94B2D7'
# Havelock Blue
color_blue2 = '#4A81BF'
# Salem Green
color_green1 = '#0C8040'
# Celery
color_green2 = '#9ABB59'

# create new figure
fig, ax1 = plt.subplots(figsize=(9.5, 7.2), # width, height in inches 
                        dpi=110)            # resolution of the figure

# tune the subplot layout by setting sides of the figure
fig.subplots_adjust(left=0.103, right=0.7, top=0.881, bottom=0.096)

# set the data limits for the y,x-axis
ax1.set_ylim([-100, 250])
ax1.set_xlim([-0.5, 5.4])
# move x axes to zero
ax1.spines['bottom'].set_position('zero')

plt.bar(list(range(0, 6)), YA, width=0.68, color=color_blue1, edgecolor=color_blue1, 
        linewidth=0.9)
plt.bar(list(range(0, 6)), YT, width=0.68, color=color_blue2, edgecolor=color_blue2, 
        linewidth=0.9)
plt.bar(list(range(0, 6)), YP, width=0.68, bottom=YT, color=color_green1, edgecolor=color_green1, 
        linewidth=0.9)
plt.bar(list(range(0, 6)), YQ, width=0.68, bottom=YT+YP, color=color_green2, edgecolor=color_green2, 
        linewidth=0.9)
plt.bar(list(range(0, 6)), YU, width=0.68, bottom=YT+YP+YQ, color='white',
        edgecolor=color_gray5, 
        linewidth=0.9)

# set properties for axes object (ticks for all issues with labels)
plt.setp(ax1, xticks=list(range(0, 6)), xticklabels=X)

# remove top and right graph borders and change bottom-left axes appearance
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_color(color_gray1)
ax1.spines['left'].set_color(color_gray1)
ax1.spines['left'].set_linewidth(1.5)
ax1.spines['bottom'].set_linewidth(1.5)

# change y ticks color
ax1.tick_params(axis='y', colors=color_gray1)
# hide x ticks
ax1.tick_params(bottom='off')

# configure x tick label appearance
for item in ax1.get_xticklabels():
    item.set_fontsize(16)
    item.set_color(color_gray4)
    # use trasformations to shift x tick labels slightly down
    offset = transforms.ScaledTranslation(0, -0.07, fig.dpi_scale_trans)
    item.set_transform(item.get_transform() + offset)
    
# configure y tick label appearance
for item in ax1.get_yticklabels():
    item.set_fontsize(16)
    item.set_color(color_gray5)
    # use trasformations to shift y tick labels slightly left
    offset = transforms.ScaledTranslation(-0.07, 0, fig.dpi_scale_trans)
    item.set_transform(item.get_transform() + offset)

# add footnote with the data source
plt.text(-0.25, -10,
         'Based on survey of 100 students conducted '
         'before and after pilot program '
         '(100% response rate on both surveys).',          
         fontsize=10.5,
         color=color_gray1)
```
