```python
import pandas as pd
from matplotlib import transforms, pyplot as plt
import numpy as np
from io import StringIO
from matplotlib.patches import Rectangle
import numpy as np
import matplotlib

%matplotlib inline
```

```python
X = ['Bored', 'Not great', 'OK', 'Kind of\ninterested', 'Excited']
Y = [11, 5, 40, 25, 19]
Z = [12, 6, 14, 30, 38]

plt.figure(figsize=(8.2, 5))

def subcategorybar(X, 
                   vals, 
                   base_colors,
                   special_colors,
                   special_bars,
                   width=0.9):
    n = len(vals)
    _X = np.arange(len(X))
    for i in range(n):
        bars = plt.bar(_X - width/2. + i/float(n)*width, 
                       vals[i], 
                       width=width/float(n), 
                       align="edge", 
                       color=base_colors[i])
        for bi, b in enumerate(bars):
            plt.text(b.get_x()+0.07, 
                     b.get_height()-3, 
                     str(vals[i][bi])+ '%', 
                     fontsize=12, 
                     color='white')
        if special_bars[i] != None:
            for special_bar_i in special_bars[i]:
                bars[special_bar_i].set_color(special_colors[i])
    plt.xticks(_X, X)

subcategorybar(X, [Y,Z], ['#A8A9AD', '#94B2D7'], ['#939598', '#4A81BF'], [[2], [3, 4]])

for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
plt.tick_params(top='off', 
                bottom='off', 
                left='off', 
                right='off', 
                labelleft='off', 
                labelbottom='on')
plt.subplots_adjust(left=0, 
                    right=0.735, 
                    top=0.67, 
                    bottom=0.15)

for item in plt.gca().get_xticklabels():
    item.set_fontsize(12)
    item.set_color('#555655')
    
plt.text(-0.4, 35, 
         '$\\bf{BEFORE}$ program, the\n'
         'majority of children felt\n' 
         'just $\\mathit{OK}$ about science.', 
         fontsize=12, 
         fontname='Arial',
         color='#555655')
plt.text(4.5, 16, 
         '$\\bf{AFTER}$\n'
         'program,\n'
         'more children\n'
         'were $\\it{Kind\ of}$\n'
         '$\\it{interested\ &}$\n'
         '$\\it{Excited}$ about\n'
         'science.', 
         fontsize=12, 
         fontname='Arial',         
         color='#4A81BF')
plt.text(-0.6, -11, 
         'Based on survey of 100 students conducted '
         'before and after pilot program '
         '(100% response rate on both surveys).', 
         color='#555655',
         fontname='Arial')
plt.text(-0.61, 50, 
         'How do you feel about science?',
         fontname='Arial',
         fontsize=19)
plt.text(-0.6, 60, 
         'Pilot program was a success'+' '*25, 
         fontsize=26, 
         fontname='Arial',
         color='white', 
         bbox={'facecolor':'#939598', 'pad':12, 'edgecolor':'none'})
```
