```python
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# configure plot font family to Arial
plt.rcParams['font.family'] = 'Arial'
# configure mathtext bold and italic font family to Arial
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'

# Battleship Grey (footnote)
color_gray1 = '#555655'
# Aluminium (compared to note)
color_gray2 = '#929497'
# Grey Chateau (title background)
color_gray3 = '#939598'
# Havelock Blue (text)
color_blue2 = '#4A81BF'

# create new figure
plt.figure(figsize=(8.2, 5), # width, height in inches
           dpi=110)          # resolution of the figure

# remove chart borders
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# change the appearance of ticks, tick labels, and gridlines (remove everything)
plt.tick_params(top='off', 
                bottom='off', 
                left='off', 
                right='off', 
                labelleft='off', 
                labelbottom='off')

# titile the plot
plt.text(-0.15, 1.03, 
         'Pilot program was a success'+' '*25, # trailing whitespaces for background
         fontsize=26, 
         color='white', 
         # put a rectangular box around the text to set a background color
         bbox={'facecolor':color_gray3, # background color
               'pad':12,                # add more padding
               'edgecolor':'none'})     # no border

# add footnote with the data source
plt.text(-0.14, -0.04,          
         'Based on survey of 100 students conducted '
         'before and after pilot program '
         '(100% response rate on both surveys).',          
         color=color_gray1)

# add note to the plot
plt.text(-0.14, 0.8, 
         'After the pilot program,', 
         fontsize=18, 
         color=color_blue2)

# add note to the plot
plt.text(-0.14, 0.38, 
         '$\\bf{68\%}$', # use mathtext \\bf for bold text
         fontsize=122, 
         color=color_blue2)

# add note to the plot
plt.text(-0.14, 0.18, 
         # use mathtext \\bf for bold text
         '$\\bf{of\ kids\ expressed\ interest\ towards\ science,}$', 
         fontsize=18, 
         color=color_blue2)

# add note to the plot
plt.text(-0.14, 0.1, 
         'compared to 44% going into the program.', 
         fontsize=18, 
         color=color_gray2)
```
